# Pre-knowledge/ Undertanding of concept:

```python
import pandas as pd
from difflib import SequenceMatcher
# Function to check if similarity is less than 0.3
def is_low_similarity(similarity):
    return similarity < 0.3

# Sample 2 easy, 2 medium, and 2 difficult questions as before
easy_questions = df[df['Difficulty'] == 'Easy'].sample(2)
medium_questions = df[df['Difficulty'] == 'Medium'].sample(2)
difficult_questions = df[df['Difficulty'] == 'Difficult'].sample(2)

# Concatenate the selected questions
questions_to_ask = pd.concat([easy_questions, medium_questions, difficult_questions])

# List to store similarity scores
similarity_scores = []

# List to store topics with low similarity
low_similarity_topics = []

# Ask and evaluate user answers for selected questions
for index, row in questions_to_ask.iterrows():
    user_answer = input(row['Question'] + "\nYour answer: ")
    actual_answer = row['Answer']
    similarity = SequenceMatcher(None, user_answer, actual_answer).ratio()
    print("Similarity score:", similarity)
    similarity_scores.append(similarity)

    # Check if similarity is low and add the topic to the list
    if is_low_similarity(similarity):
        low_similarity_topics.append(row['Topic'])

print("Similarity scores:", similarity_scores)

# Find unique topics with low similarity
unique_low_similarity_topics = set(low_similarity_topics)

# Print the topics with low similarity
print("Topics with Low Similarity Scores:", unique_low_similarity_topics)
```

# Round 1 setting Bayesian probability threshold:

```python
columns = ['Question', 'Answer','Topic','Difficulty']
empty_df1 = pd.DataFrame(columns=columns)
# Iterate through topics with low similarity
for topic in unique_low_similarity_topics:
    # Sample one random question along with its answer from the current topic
    topic_question = df[df['Topic'] == topic].sample(10)

    # Append the selected question and answer to the additional_questions list
    empty_df1 = empty_df1.append(topic_question, ignore_index=True)
easy_questions = empty_df1[empty_df1['Difficulty'] == 'Easy'].sample(2)
medium_questions = empty_df1[empty_df1['Difficulty'] == 'Medium'].sample(2)
difficult_questions = empty_df1[empty_df1['Difficulty'] == 'Difficult'].sample(2)

# Concatenate the selected questions
questions_to_ask = pd.concat([easy_questions, medium_questions, difficult_questions])

# List to store similarity scores
similarity_scores = []

# List to store topics with low similarity
low_similarity_topics = []

# Ask and evaluate user answers for selected questions
for index, row in questions_to_ask.iterrows():
    user_answer = input(row['Question'] + "\nYour answer: ")
    actual_answer = row['Answer']
    similarity = SequenceMatcher(None, user_answer, actual_answer).ratio()
    print("Similarity score:", similarity)
    similarity_scores.append(similarity)

    # Check if similarity is low and add the topic to the list
    if is_low_similarity(similarity):
        low_similarity_topics.append(row['Topic'])

print("Similarity scores:", similarity_scores)
```

# Round 3 of adaptive learning:

```python
import pandas as pd
from difflib import SequenceMatcher

# Define thresholds for each difficulty level
thresholds = {
    'Easy': 0.1,
    'Medium': 0.1,
    'Difficult': 0.1
}

# Initialize a list to store results
results = []

# Iterate through questions and evaluate answers
for index, row in questions_to_ask.iterrows():
    user_answer = input(row['Question'] + "\nYour answer: ")
    actual_answer = row['Answer']
    similarity = SequenceMatcher(None, user_answer, actual_answer).ratio()

    # Determine the difficulty level of the question
    difficulty_level = row['Difficulty']

    # Get the topic of the question
    topic = row['Topic']

    # Determine if the answer is right or wrong based on similarity
    if similarity < 0.2:
        result = 'Wrong'
    else:
        result = 'Right'

    # Append the result to the list
    results.append({'Question': row['Question'],
                    'Difficulty': difficulty_level,
                    'Topic': topic,
                    'Similarity': similarity,
                    'User Answer': user_answer,
                    'Actual Answer': actual_answer,
                    'Result': result})

# Create a DataFrame from the results list
result_df = pd.DataFrame(results)

# Print the result DataFrame
result_df.head()
```
